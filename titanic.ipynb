```python
import pandas as pd
df = pd.read_csv('titanic.csv')
df.head(10)
```

```python
input = df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns')
input.head(10)
```

```python
target = df['Survived']
```

```python
import math
median_age = math.floor(df.Age.median())
median_age
```

```python
input.Age = input.Age.fillna(median_age)
input.head(10)
```

```python
from sklearn.preprocessing import LabelEncoder
le_Sex = LabelEncoder()

input['Sex_n'] = le_Sex.fit_transform(input['Sex'])
inputs_new = input.drop(['Sex','Survived'],axis='columns')
inputs_new.head(10)
```

```python
target
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_new,target)
```

```python
model.score(inputs_new,target)
```

```python
model.predict([[1,26,8,0]])
```
